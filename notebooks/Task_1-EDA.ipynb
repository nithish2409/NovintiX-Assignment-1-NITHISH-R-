```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
sns.set_style("whitegrid")
warnings.filterwarnings('ignore')

print("EDA notebook started...")
```

```python
file_path = '../data/healthcare_dataset.csv'
try:
    df = pd.read_csv(file_path)
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
```

```python
print(df.head(55500))
print("\n--- Data Info ---")
df.info()
```

```python
print("\n--- Missing Values ---")
print(df.isnull().sum())
```

```python
# (Age, Billing Amount, and Room Number)

print("\n--- Task 1.1: Analyzing Numerical Distributions ---")

# Get a statistical summary (mean, median, min, max, quartiles)
numerical_cols = ['Age', 'Billing Amount', 'Room Number']
print("Statistical Summary for Numerical Columns:")
print(df[numerical_cols].describe())
```

```python
# Visualize the distributions using histograms
plt.figure(figsize=(20, 6))

# Plot for Age
plt.subplot(1, 3, 1)
sns.histplot(df['Age'], bins=30, kde=True, color='blue')
plt.title('Distribution of Age', fontsize=16)
plt.xlabel('Age')
plt.ylabel('Frequency')

# Plot for Billing Amount
plt.subplot(1, 3, 2)
sns.histplot(df['Billing Amount'], bins=30, kde=True, color='green')
plt.title('Distribution of Billing Amount', fontsize=16)
plt.xlabel('Billing Amount')
plt.ylabel('Frequency')

# Plot for Room Number
plt.subplot(1, 3, 3)
sns.histplot(df['Room Number'], bins=30, kde=True, color='purple')
plt.title('Distribution of Room Number', fontsize=16)
plt.xlabel('Room Number')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# (Medical Condition, Admission Type, and Medication)

print("\n--- Task 1.2: Visualizing Categorical Frequencies ---")
```

```python
# Define the columns to plot
categorical_cols = ['Medical Condition', 'Admission Type', 'Medication']

# Create count plots for each categorical variable
# Using horizontal bar charts (countplot with y) is better for readability
plt.figure(figsize=(20, 18))

# Plot for Medical Condition
plt.subplot(3, 1, 1)
sns.countplot(y=df['Medical Condition'], order=df['Medical Condition'].value_counts().index, palette='viridis')
plt.title('Frequency of Medical Condition', fontsize=16)
plt.xlabel('Count')
plt.ylabel('Medical Condition')

# Plot for Admission Type
plt.subplot(3, 1, 2)
sns.countplot(y=df['Admission Type'], order=df['Admission Type'].value_counts().index, palette='plasma')
plt.title('Frequency of Admission Type', fontsize=16)
plt.xlabel('Count')
plt.ylabel('Admission Type')

# Plot for Medication
plt.subplot(3, 1, 3)
sns.countplot(y=df['Medication'], order=df['Medication'].value_counts().index, palette='cividis')
plt.title('Frequency of Medication', fontsize=16)
plt.xlabel('Count')
plt.ylabel('Medication')

plt.tight_layout()
plt.show()

print("\nEDA Task 1 Complete.")
```
